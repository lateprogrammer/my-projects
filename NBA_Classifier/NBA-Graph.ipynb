```python
%matplotlib inline

import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nba = pd.read_csv('test_data.csv')
```

```python
nba.head()
```

```python
# Declare colors
silver = '#989898'
drk_gray = '#555555'
cardinals_red = '#C41E3A'
blues_drk = '#0f1f48'
blues_lt = '#0447a0'
blues_gold = '#ffc325'

# Create a new figure
plt.figure(figsize=(20, 10), dpi=100)


# plt.xticks(nba['name'], rotation='vertical')

plt.tick_params(
    axis='both', 
    which='major', 
    labelsize=14, 
    colors=silver)

axes = plt.gca()
axes.spines['top'].set_visible(False)
axes.spines['right'].set_visible(False)
axes.spines['left'].set_visible(False)
axes.spines['bottom'].set_visible(False)

plt.scatter(
    nba['probability_predictions'], 
    nba['gp'], 
    c = cardinals_red, s=100, label = 'test_data')

plt.xlabel('Test_Data: Prediction Probability',
           fontsize=18,
           color=drk_gray)
plt.ylabel('Test_Data: Games Played',
           fontsize=18,
           color=drk_gray)


plt.title('Test_Data: Games Playes VS Prediction Proability',
    fontsize=25,
    color=drk_gray)
plt.suptitle(
    'NBA ROOKIE PREDICTIONS', 
    y = 1, 
    color = drk_gray,
    fontsize = 18)

axes.grid(
    axis='both', 
    color=silver, 
    linestyle='dotted', 
    linewidth=1,
    zorder=1)
leg = axes.legend(loc='lower right', prop={'size': 16})
ax = plt.gca()
ax.axis([0,1,0,100])

fig = plt.gcf()
fig.savefig('NBA-ROOKIE.png', dpi=100)
```
